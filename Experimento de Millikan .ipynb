```python
#Constante de Planck
import matplotlib.pyplot as plt
import numpy as np
#cte de planck = 6,62607015 × 10-34 (m^2 kg / s)
#h = constante de Planck, e a carga do elétron ,t a função trabalho do materia, frequência f e potencial V
dados = np.loadtxt('millikan.txt') #Lê o arquivo
lista_dados = np.array(dados) #Cria lista do arquivo

f = lista_dados[:,0]
v = lista_dados[:,1]
x = f
y = v

plt.scatter(x,y, color='black')
plt.title("Experimento de Millikan")
plt.xlabel(R"Frequência $v$ em $[Herts]$")
plt.ylabel(R"Potencial $V$ em $[Volts]$")
plt.grid()
plt.show()


```

```python
dados = np.loadtxt('millikan.txt') #Lê o arquivo
lista_dados = np.array(dados) #Cria lista do arquivo

#Pontos
f = lista_dados[:,0] #Frequência
v = lista_dados[:,1] #Potencial

#Coordenadas
x = f
y = v

#Função
Funcao = np.poly1d(np.polyfit(x,y,1))
print(Funcao)

plt.plot(x,Funcao(x), color='red')
plt.scatter(x,y, color='black')
plt.title("Experimento de Millikan")
plt.xlabel(R"Frequência $v$ em $[Herts]$")
plt.ylabel(R"Potencial $V$ em $[Volts]$")
plt.grid()
plt.show()
```
